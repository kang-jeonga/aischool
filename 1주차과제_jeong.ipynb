이 점에서 STT 기술은 실로 놀라운 일이다.

















# **3.이미지**  

우리는 계절이 바뀔때마다 날씨가 맞는, 혹은 유행하는 옷을 구입하는 경우가 많다. 

물론, 지금은 온라인쇼핑으로 내일 바로 받아볼 수도 있고, 당장 픽업해갈 수 있다.

그러나 실물이 내가 기대했던 것과 다를 때가 많다는 것이 문제다.

직접 입어보고 느껴보고 하는 쇼핑을 좋아하는 경우도 있지만, 입어보는 것은 대단히 귀찮아하는 사람도 많을 것이다.
<br/>
<br/>
>- **인공지능 디자이너**  

<img src = "http://image.chosun.com/sitedata/image/201807/27/2018072701968_1.jpg" >



흥미로운 점은, 직접 입어보지 않고 인공지능을 이용해 스타일링을 할 수 있다.  

즉, 인공지능이 패션을 이해하고  학습하며,이미지 인식 기술을 접목한것이 패션 AI라 할수있다.  

이미지 인식 기술은 SNS 등에서 방대한 패션관련 데이터를 사용해  

요즘의 트랜드, 핫한 이슈, 의류소재의 정보등 중요한 자료를 추출하여, 각각의 사용자의 취향에 맞는 스타일을 추천해준다.  
<br/>
<br/>


>- **비지도 학습 GAN (Generative Adversarial Networks)**

패션AI, 즉  AI디자이너는 수많은 데이터를 추출은 기본이고,그 중 유니크한 점은 비지도학습이라 생각해본다.  

머신러닝은 컴퓨터가 데이터를 학습해 패턴을 찾아내는 알고리즘으로 지도학습,비지도학습,강화학습으로 분류된다.  

지도학습은 정답이 주어진 상태로 학습하며, 비지도학습은 문제는 주되, 정답을 알려주지않고 학습시킨다.  

패션AI의 중요 기술은 바로 [비지도학습 GAN](https://www.samsungsds.com/global/ko/support/insights/Generative-adversarial-network-AI-2.html) 이다.


이는 능동적으로 창작물을 생성하고, 스스로 이미지를 만들기도 하고,
**_이미지가 진짜인지 가짜인지도 구분_** 하기도 한다.  

이로써, 사용자의 이미지를 보고 패션 조언을 해주며, **_유행을 예측_** 해 새로운 스타일을 생성하기도 한다.  














 


#**4. 자율주행**  


자동차 부품을 생산하는 회사에서 근무했던 나는, 납품을 하는데 있어서 필수인 지입차량관련일들도 종종 봐왔었다.   

영업부 직원이 푸념조로 무인트럭이 곧 나올거라고 했었는데, 그건 먼 미래일거라 넘겨 들었는데 생각해보기도 전에 상용화가 멀지 않은 듯 하다.

<br/>


>- **무인 자율주행트럭**

납기준수는 필수적이다. 납품담당기사님들 모두 전문적이고 경력도 많지만, 사람인지라 졸음운전이나 사고등을  배제할수는 없다.  


무인자율주행트럭이 가능하다면 이동시간을 단축시키고,  트럭운전자의 운전보다 안전 할 것이다. 
<br/>
<br/> 



>- **인공신경망 CNN**


자율주행 자동차의 작동원리는 '인지->판단->제어'의 3단계를 거친다.  

인지단계에서는 카메라등의 센서로 주변환경을 인지하고,  

판단단계에서는 그 정보를 해석하여 안전한 주행이 가능하도록 주행경로를 생성한다.  

마지막으로 제어단계는 그 경로를 따르는 가감속 조향제어를 한다.  

이렇게 복잡한 일을 수행하기 위해서는 고도의 알고리즘, 고성능컴퓨터가 필요하다.  

이런 복잡한 기술들 주목할 만한 기술은 **_인간의 시신경 작동 원리를 모방_** 한  

[CNN(Convolutional neural network)](https://kr.mathworks.com/solutions/deep-learning/convolutional-neural-network.html)이라 불리는 컨볼루션 신경망, 즉 인공신경망이다.  

이 기술은 딥러닝의 일종으로 학습을 거듭하면서 영상내에 중요한 점들을 추출하여  

특징 맵(feature map)을 적절하게 생성한다.
